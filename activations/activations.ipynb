```python
from activations import *
import matplotlib.pyplot as plt
```

```python
def plot_activation(activation_fn, x_min=-5, x_max=5, num=1000, title=None):
    """
    Plot any scalar, NumPy-vectorised activation function.

    Parameters
    ----------
    activation_fn : callable
        Function applied element-wise (e.g., relu, np.tanh).
    x_min, x_max : float
        Range of inputs to visualise.
    num : int
        Number of sample points.
    title : str or None
        Custom plot title.
    """
    x = np.linspace(x_min, x_max, num)
    y = activation_fn(x)

    plt.figure()
    plt.plot(x, y)  # default styling
    plt.xlabel("Input")
    plt.ylabel("Output")
    plt.title(title or activation_fn.__name__)
    plt.grid(True, linestyle='--', linewidth=0.5)
    plt.show()
```

```python
plot_activation(ReLU)
```

```python
plot_activation(PReLU)
```
